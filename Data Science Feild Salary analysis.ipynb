# Web Scraping

```python
from bs4 import BeautifulSoup as bs
import re
import requests
```

```python
url="https://countrycode.org/"
req=requests.get(url)
req
```

```python
soup=bs(req.content)
soup
```

```python
soup.findAll("td")
```

# Regex pattern for finding country and its code

```python
country=[]
code=[]
count=0
for i in soup.findAll("td"):
    if count==6:
        count=0
    if count==0:
        val=re.sub('^<td>.+">|<.*>',"",str(i))
        country.append(val)
    elif count==2:
        val=re.sub('<td>|/.*</td>',"",str(i))
        code.append(val.rstrip())
    count+=1    
```

# Data Cleaning

```python
df.drop(["salary","salary_currency"],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df["experience_level"]=df["experience_level"].map({"EN":"Entry-level","SE":"Senior-level",
                                                   "EX":"Executive-level","MI":"Mid-level"})
df["employment_type"]=df["employment_type"].map({"FT":"Full-time","CT":"Contract","PT":"Part-time","FL":"Freelance"})
df["company_size"]=df["company_size"].map({"S":"Small","L":"Large","M":"Medium"})
df["employee_residence"]=df["employee_residence"].map(dict(zip(code,country)))
df["company_location"]=df["company_location"].map(dict(zip(code,country)))
```

# Data Visualization

```python
plt.figure(figsize=(15,40))
for i,j in enumerate(df["work_year"].unique()):
    plt.subplot(5,1,i+1)
    ax=sns.countplot(x=df[df["work_year"]==j]["company_location"])
    for bars in ax.containers:
        ax.bar_label(bars)
    plt.title("Number of companies in each state in the year "+str(j),fontsize=15,fontdict={"color":"green"})
    plt.xlabel("Number of companies in each country",fontsize=15,fontdict={"color":"blue"})
    plt.ylabel("count of companies",fontsize=15,fontdict={"color":"red"})
    plt.xticks(rotation=90)
plt.tight_layout(h_pad=3)    
plt.show()    
```

```python
comp_size=df.groupby(["work_year","company_size"])["job_title"].value_counts().reset_index().sort_values(by=["company_size","count"],ascending=[True,False])
comp_size
```

```python
fig,axes=plt.subplots(5,3,figsize=(30,43))
for i,year in enumerate(comp_size["work_year"].unique()):
    for j,size in enumerate(comp_size["company_size"].unique()):
        new=comp_size[(comp_size["work_year"] == year) & (comp_size["company_size"] == size)]
        num=sns.barplot(x=new["job_title"].head(),y=new["count"].head(),ax=axes[i,j])
        for bars in num.containers:
            num.bar_label(bars)
        axes[i,j].set_title("Top 5 job roles in "+size+" company in "+str(year),fontsize=20,fontdict={"color":"green"})
        axes[i,j].set_xlabel("Job title",fontsize=20,fontdict={"color":"blue"})
        axes[i,j].set_ylabel("Count",fontsize=20,fontdict={"color":"red"})
plt.tight_layout(h_pad=2)
plt.show()
```

```python
plt.figure(figsize=(15,15))
c1=["#228B22","#FF8C00","#FFD700","#B22222"]
c2=["#87CEEB","#00FFFF","#008080","#20B2AA"]
for i,j in enumerate(df["work_year"].unique()):
    emp_level=df[df["work_year"]==j]["experience_level"].to_frame()["experience_level"].map({"Entry-level":1,"Mid-level":2,"Senior-level":3,"Executive-level":4}).value_counts().to_frame().reset_index()
    emp_level["label"]=emp_level["experience_level"].map({1:"Entry-level",2:"Mid-level",3:"Senior-level",4:"Executive-level"})
    plt.subplot(3,2,i+1)
    if i%2==0:
        c=c1
    else:
        c=c2
    plt.pie(x=emp_level["count"],labels=dict(zip(emp_level["label"],emp_level["experience_level"])),autopct="%1i%%",colors=c,wedgeprops={"edgecolor":"black"})
    plt.pie([1],colors="white",radius=0.7,wedgeprops={"edgecolor":"black"})
    plt.title("Trends of experience level in the year "+str(j))
plt.show()    
```

```python
plt.figure(figsize=(15,20))
for i,j in enumerate(df["work_year"].unique()):
    plt.subplot(5,1,i+1)
    vs=df[df["work_year"]==j]
    ax=sns.countplot(x=vs["experience_level"],hue=vs["employment_type"])
    for bars in ax.containers:
        ax.bar_label(bars)
    plt.title("Employment type in each experience level in "+str(j),fontsize=15,fontdict={"color":"green"})
    plt.xlabel("experience_level",fontsize=15,color="orange")
    plt.ylabel("Count",fontsize=15,color="red")
plt.tight_layout()    
plt.show()    
```

```python
col=["Reds","Greens","Oranges","Purples","Blues"]
plt.figure(figsize=(8,15))
for i,j in enumerate(df["work_year"].unique()):
    plt.subplot(5,1,i+1)
    top_5_sal=df.groupby(["work_year","job_title"])["salary_in_usd"].max().reset_index().sort_values(by=["work_year","salary_in_usd"],ascending=[True,False])
    ax=sns.barplot(x=top_5_sal[top_5_sal["work_year"]==j]["salary_in_usd"].head(),y=top_5_sal[top_5_sal["work_year"]==j]["job_title"].head(),orient="horizontal",palette=col[i])
    for bars in ax.containers:
        ax.bar_label(bars,padding=-40,color="white")
    plt.xticks([0,200000,400000,600000,800000])
    plt.title("Top 5 Highest paid jobs in "+str(j),fontsize=15,fontdict={"color":"green"})
    plt.xlabel("Salary in USD",fontsize=15,fontdict={"color":"orange"})
    plt.ylabel("Job title",fontsize=15,fontdict={"color":"red"})
plt.tight_layout()
plt.show()
```

```python
max_sal=df.groupby(["experience_level","work_year"])["salary_in_usd"].max().reset_index()
max_sal
```

```python
plt.figure(figsize=(10,15))
for i in max_sal["experience_level"].unique():
    year=max_sal[max_sal["experience_level"]==i]
    sns.lineplot(x=year["work_year"],y=year["salary_in_usd"],label=i,marker="D")  
    for j in range(len(year)):
        plt.text(year["work_year"].iloc[j]+0.03,year["salary_in_usd"].iloc[j],year["salary_in_usd"].iloc[j],fontsize=9)
plt.xticks([2020,2021,2022,2023,2024]) 
plt.title("MAX salary for experience level in each year",fontsize=15,fontdict={"color":"green"})
plt.xlabel("Work year",fontsize=15,fontdict={"color":"orange"})
plt.ylabel("Salary in USD",fontsize=15,fontdict={"color":"red"})
plt.grid()
plt.show()
```
